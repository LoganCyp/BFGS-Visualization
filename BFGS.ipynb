```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rosenbrock(x):
    f = 100*(x[0]**2-x[1])**2+(x[0]-1)**2
    return f

def f(x,y):
    f = 100*(x**2-y)**2+(x-1)**2
    return f

def path(x):
    global xs
    xs.append(x)
    
x0 = np.array([5,3])
xs = [x0]

v_func = np.vectorize(f) 

x, y = np.meshgrid(np.linspace(0, 10, 100),
                   np.linspace(-3, 40, 100))
```

```python
result = minimize(rosenbrock, x0, callback=path)
print(result)

xs = np.array(xs)

fig, ax = plt.subplots(1)
plt.plot(xs[:,0],xs[:,1],'g-o')
CS = ax.contour(x, y, v_func(x, y))
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('BFGS')
plt.show()
```
